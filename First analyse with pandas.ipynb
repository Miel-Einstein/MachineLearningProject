```python
def multiple_unique(liste):
    if len(liste)==0:
        result=np.nan
    elif len(liste)==1:
        result="unique"
    else :
        result="multiple"
    return result
```

```python
data['nombresStyles']=multiple_unique(data['style'])
```

```python
data['nombresStyles'].value_counts()
```

```python
data['nombresStyles']=data['style'].apply(lambda x : multiple_unique(x))
```
